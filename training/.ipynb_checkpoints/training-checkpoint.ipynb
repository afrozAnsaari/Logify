```python
import pandas as pd
```

```python
df=pd.read_csv("./dataset/raw_synthetic_logs.csv")
```

```python
df.head(3)
```

```python
df.source.unique()
```

```python
df.to_csv("raw_synthetic_logs.csv", index=False)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import DBSCAN

model=SentenceTransformer("all-MiniLM-L6-v2")

embeddings=model.encode(df["log_message"].to_list())

embeddings[:2]
```

```python
len(embeddings[0])
```

```python
dbscan=DBSCAN(eps=0.2, min_samples=1, metric="cosine")
clusters=dbscan.fit_predict(embeddings)


df['cluster']=clusters
```

```python
df['cluster'].unique()
```

```python
df.head()
```

```python
df[df.cluster==1]
```

```python
cluster_counts=df['cluster'].value_counts()
large_cluster=cluster_counts[cluster_counts>10].index

print(f"Cluster Count:{cluster_counts}\n")

print(f"Large Cluster: {large_cluster}\n")

for cluster in large_cluster:
    print(f"Cluster: {cluster}")
    print(df[df['cluster']==cluster]['log_message'].head(5).to_string(index=False))
    print()
```

```python
import re
def classify_with_regex(log_message:str):
    regex_patterns = {
        r"User User\d+ logged (in|out).": "User Action",
        r"Backup (started|ended) at .*": "System Notification",
        r"Backup completed successfully.": "System Notification",
        r"System updated to version .*": "System Notification",
        r"File .* uploaded successfully by user .*": "System Notification",
        r"Disk cleanup completed successfully.": "System Notification",
        r"System reboot initiated by user .*": "System Notification",
        r"Account with ID .* created by .*": "User Action"
    }
    for pattern, label in regex_patterns.items():
        if re.search(pattern, log_message, re.IGNORECASE):
            return label
    return None    
```

```python
classify_with_regex("User User494 logged out.")
```

```python
df["regex_label"]=df['log_message'].apply(classify_with_regex)
df[df.regex_label.notnull()].shape
```

```python
df_non_regex=df[df['regex_label'].isnull()].copy()
df_non_regex.head(4)
```

```python
print(df_non_regex['target_label'].value_counts()[df_non_regex['target_label'].value_counts()<=5].index.tolist())
```

```python
df_non_legacy_regex=df_non_regex[df_non_regex.source!="LgeacyCRM"]
df_non_legacy_regex.source.unique()
```

```python
model=SentenceTransformer("all-MiniLM-L6-v2")

filtered_embeddings=model.encode(df_non_legacy_regex["log_message"].to_list())

filtered_embeddings[:2]
```

```python
X=filtered_embeddings
y=df_non_legacy_regex['target_label']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=42)
clf=LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
report=classification_report(y_test, y_pred, zero_division=0)
print(report)
```

```python
import joblib
joblib.dump(clf, "models/log_classifier.joblib")
```
